```python
writer = json_rec.ParamWriter()
param = writer.read(file_dir="/home/ec2-user/ga/gen6/rank0_0.02574.json")
```

```python
param["num_neurons"] = [256, 256, 512, 512]
param["LAGS"] = 32
param["epochs"] = 1000
param["learning_rate"] = 0.01
```

```python
# param = {'PERIODS': 1, 'ORDER': 1, 'LAGS': 5, 'TRAIN_RATIO': 0.8, 'VAL_RATIO': 0.1, 'epochs': 500, 'num_inputs': 1, 'num_outputs': 1, 'num_time_steps': 5, 'num_neurons': (94, 225), 'learning_rate': 0.009794101837580605, 'clip_grad': None, 'report_periods': 10, 'tensorboard_path': '/home/ec2-user/ec2_hps/2018DEC21_01/tensorboard/', 'model_path': '/home/ec2-user/ec2_hps/2018DEC21_01/saved_models/', 'fig_path': '/home/ec2-user/ec2_hps/2018DEC21_01/model_figs/'}
```

```python
df_ready = rnn_prepare.prepare_dataset(
    file_dir=FILE_DIR,
    periods=int(param["PERIODS"]),
    order=int(param["ORDER"]),
    remove=".",
    verbose=True
)

# Split dataset.
(X_train, X_val, X_test,
 y_train, y_val, y_test)\
= rnn_prepare.split_dataset(
    raw=df_ready,
    train_ratio=param["TRAIN_RATIO"],
    val_ratio=param["VAL_RATIO"],
    lags=param["LAGS"]
)

# The gross dataset excluding the test set.
# Excluding the test set for isolation purpose.
data_feed = {
    "X_train": X_train,
    "X_val": X_val,
    "y_train": y_train,
    "y_val": y_val,
}
ep = param["epochs"]
ckpts = range(int(ep * 0.95), ep)  # Take the final 5% epochs.
net = stacked_lstm.StackedLSTM(
    param=param,
    prediction_checkpoints=ckpts,
    verbose=True
)
ret = net.fit(
    data=data_feed,
    ret="all",
    save_to_disk=False
)
```

```python
step = param["epochs"] - 1
plt.close()
plt.style.use('seaborn-dark')
rg_train = range(len(data_feed["y_train"]))
rg_val = range(len(data_feed["y_train"]),
               len(data_feed["y_train"]) + len(data_feed["y_val"]))
plt.figure(figsize=(256, 8))

plt.plot(rg_train, ret["pred_all"][step]["train"])
plt.plot(rg_train, data_feed["y_train"])
plt.plot(rg_val, ret["pred_all"][step]["val"])
plt.plot(rg_val, data_feed["y_val"])

plt.grid(True)
plt.legend(["pred_train", "actual_train", "pred_val", "actual_val"], loc="best")
plt.show()
```
